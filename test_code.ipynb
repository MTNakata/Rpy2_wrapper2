```python
import pandas as pd
import rpy2.robjects as ro
from rpy2.robjects import pandas2ri
from rpy2_wrapper2 import *

# R package requirements: "DescTools"
# Please install this package in R console: install.packages("DescTools")
```

```python
# Import iris dataset from R and transfer to Pandas DataFrame

ro.r("data(iris)")
ro.r("iris$Species <- as.character(iris$Species)")
iris = pandas2ri.rpy2py(ro.r["iris"])
iris.head() 
```

```python
# Import iris data from Pandas DataFrame and transfer to R

data = "new_df"
ro.globalenv[data] = pandas2ri.py2rpy(iris)
```

```python
category = "Species"
variables = ["Sepal.Length","Sepal.Width","Petal.Length","Petal.Width"]

tukey_results = []

for variable in variables:
    result = tukeyHSD(df="iris",category=category,variable=variable)
    tukey_results.append(result)

tukey_results_df = pd.concat(tukey_results)
tukey_pvalues = tukey_results_df[tukey_results_df["params"]=="p.adj"]
tukey_pvalues
```

```python
control="setosa"

dunnett_results = []

for variable in variables:
    result = dunnett_test(df=data,category=category,variable=variable,control=control)
    dunnett_results.append(result)

dunnett_results_df = pd.concat(dunnett_results)
dunnett_pvalues = dunnett_results_df[dunnett_results_df["params"]=="pval"]
dunnett_pvalues
```

```python
rpy2_close()
```
